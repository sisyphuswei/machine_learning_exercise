```python
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

np.random.seed(8)
x1 = np.random.normal(scale=2, size=(100,1))
x2 = np.random.normal(scale=1, size=(100,1))
X = np.hstack((x1,x2))

# keep the first n principal components of the data 
pca = PCA(n_components=1)
# fit PCA model to rotX
pca.fit(X)
# transform data onto the first two principal components
X_pca = pca.transform(X)
print("PCs are {}".format(pca.components_))

plt.figure(figsize=(8, 8))
plt.scatter(X[:,0], X[:,1])
plt.xlim(-7.5, 7.5)
plt.ylim(-7.5, 7.5)
plt.xlabel("$X_1$")
plt.ylabel("$X_2$")

# plot the first principal component direction
ax = plt.gca()
x_vals = np.array(ax.get_xlim())
y_vals = pca.components_[0,1]/pca.components_[0,0]*x_vals
plt.plot(x_vals, y_vals, c="y");
ax.set_aspect("equal")
plt.title("An artificial dataset")
plt.show()
```

```python
#from sklearn.datasets import load_breast_cancer 
from sklearn.preprocessing import StandardScaler
import mglearn
#cancer = load_breast_cancer()
scaler = StandardScaler() 
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python


plt.figure()
plt.scatter(X_scaled[:,0], X_scaled[:,1])
plt.axis("square")
plt.show()
```

```python
scaled_pca = PCA(n_components=2)
# fit PCA model to rotX
scaled_pca.fit(X_scaled)
# transform data onto the first two principal components
X_scaled_pca = scaled_pca.transform(X_scaled)


plt.figure()
plt.scatter(X_scaled_pca[:,0], X_scaled_pca[:,1])
plt.axis("square")
plt.show()
print(scaled_pca.components_)
print(scaled_pca.explained_variance_)
```

```python
import math

rot = math.sqrt(2)*np.array([[1, -1], [1, 1]])
rotX = (rot@X.T).T

plt.figure()
plt.scatter(rotX[:,0], rotX[:,1])
plt.axis("square")
plt.show()
```

```python
scaler = StandardScaler() 
scaler.fit(rotX)
rotX_scaled = scaler.transform(rotX)
plt.figure()
plt.scatter(rotX_scaled[:,0], rotX_scaled[:,1])
plt.axis("square")
plt.title("scaled")
plt.show()
```

```python
from sklearn.decomposition import PCA
# keep the first n principal components of the data 
rot_scaled_pca = PCA(n_components=2)
# fit PCA model to rotX
rot_scaled_pca.fit(rotX_scaled)
# transform data onto the first two principal components
rotX_pca = rot_scaled_pca.transform(rotX_scaled)

plt.figure()
plt.scatter(rotX_pca[:,0], rotX_pca[:,1])
plt.axis("square")
plt.xlabel("First principal component")
plt.ylabel("Second principal component")
plt.show()
print(rot_scaled_pca.components_)
print(rot_scaled_pca.explained_variance_)
```
