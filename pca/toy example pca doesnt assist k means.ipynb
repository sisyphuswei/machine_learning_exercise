```python
import numpy as np
import math
import matplotlib.pyplot as plt

np.random.seed(0)
x_cluster_right = np.random.normal(loc=4, scale=1, size=(100,1))
y_cluster_right = np.random.normal(scale=5, size=(100,1))
x_cluster_left = np.random.normal(loc=-4, scale=1, size=(100,1))
y_cluster_left = np.random.normal(scale=5, size=(100,1))

cluster_right = np.hstack((x_cluster_right, y_cluster_right))
cluster_left = np.hstack((x_cluster_left, y_cluster_left))

X = np.vstack((cluster_right, cluster_left))

plt.figure(figsize=(8, 8))
plt.scatter(X[:,0], X[:,1])
plt.xlim(-15,15)
plt.gca().set_aspect("equal")
plt.xlabel("$x$")
plt.ylabel("$y$")
```

```python
from sklearn.decomposition import PCA
# keep the first n principal components of the data 
pca = PCA(n_components=1)
# fit PCA model to rotX
pca.fit(X)

print("The first PC is {}".format(pca.components_))

plt.figure(figsize=(8, 8))
plt.xlim(-15,15)
plt.ylim(-15,15)
plt.scatter(X[:,0], X[:,1])
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.gca().set_aspect('equal')
ax = plt.gca()
x_vals = np.array(ax.get_xlim())
y_vals = pca.components_[0,1]/pca.components_[0,0]*x_vals
plt.plot(x_vals, y_vals, c="y");

```
