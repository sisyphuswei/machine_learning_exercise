```python
# http://archive.ics.uci.edu/ml/datasets/seeds
# The shape of seeds data is 210*7
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import mglearn
import numpy as np
import pandas as pd
np.random.seed(0)

seeds = np.loadtxt("seeds.txt")

data = seeds[:, :7]
target = seeds[:, 7]
print("The shape of seeds: {}".format(data.shape))
```

```python
display(target)
```

```python
display(data)
```

```python
from sklearn.decomposition import PCA
# keep the first two principal components of the data 
unscaled_pca = PCA(n_components=0.85, svd_solver = 'full')
# fit PCA model to breast cancer data 
unscaled_pca.fit(data)
# transform data onto the first two principal components
unscaledX_pca = unscaled_pca.transform(data)
print("Original shape: {}".format(str(data.shape))) 
print("Reduced shape: {}".format(str(unscaledX_pca.shape)))
print("Explained variance ratio: {}".format(unscaled_pca.explained_variance_ratio_))

plt.figure(figsize=(8, 8))
mglearn.discrete_scatter(unscaledX_pca[:, 0], unscaledX_pca[:, 1], target) 
#plt.legend(, loc="best") 
plt.gca().set_aspect("equal")
plt.xlabel("First principal component")
plt.ylabel("Second principal component")
```

```python
scaler = StandardScaler() 
scaler.fit(data)
data_scaled = scaler.transform(data)
#num_class = 2
pca = PCA(n_components=0.85, svd_solver = 'full')

pca.fit(data_scaled)
# transform data onto the first two principal components
X_pca = pca.transform(data_scaled)
print("Original shape: {}".format(str(data.shape))) 
print("Reduced shape: {}".format(str(X_pca.shape)))
print("Explained variance ratio: {}".format(pca.explained_variance_ratio_))

plt.figure(figsize=(8, 8))
mglearn.discrete_scatter(X_pca[:, 0], X_pca[:, 1], target) 
#plt.legend(, loc="best") 
plt.gca().set_aspect("equal")
plt.xlabel("First principal component")
plt.ylabel("Second principal component")
```
